```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
# samp = pd.read_csv('sample_submission.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
submit = test[['user_id']]
```

```python
train.drop('user_id',axis = 1, inplace = True)
test.drop('user_id',axis = 1, inplace = True)

train['REGION'].fillna(train['REGION'].mode()[0], inplace = True)
test['REGION'].fillna(test['REGION'].mode()[0], inplace = True)

train['MONTANT'].fillna(-999, inplace = True)
test['MONTANT'].fillna(-999, inplace = True)

train['FREQUENCE_RECH'].fillna(-999, inplace = True)
test['FREQUENCE_RECH'].fillna(-999, inplace = True)

train['REVENUE'].fillna(-999, inplace = True)
test['REVENUE'].fillna(-999, inplace = True)

train['ARPU_SEGMENT'].fillna(-999, inplace = True)
test['ARPU_SEGMENT'].fillna(-999, inplace = True)

train['FREQUENCE'].fillna(-999, inplace = True)
test['FREQUENCE'].fillna(-999, inplace = True)

train['DATA_VOLUME'].fillna(-999, inplace = True)
test['DATA_VOLUME'].fillna(-999, inplace = True)

train['ON_NET'].fillna(-999, inplace = True)
test['ON_NET'].fillna(-999, inplace = True)

train['ORANGE'].fillna(-999, inplace = True)
test['ORANGE'].fillna(-999, inplace = True)

train['TIGO'].fillna(-999, inplace = True)
test['TIGO'].fillna(-999, inplace = True)

train['ZONE1'].fillna(-999, inplace = True)
test['ZONE1'].fillna(-999, inplace = True)

train['ZONE2'].fillna(-999, inplace = True)
test['ZONE2'].fillna(-999, inplace = True)

train.drop('MRG',axis = 1, inplace = True)
test.drop('MRG',axis = 1, inplace = True)

train['FREQ_TOP_PACK'].fillna(-999, inplace = True)
test['FREQ_TOP_PACK'].fillna(-999, inplace = True)

train['TOP_PACK'].fillna(train['TOP_PACK'].mode()[0], inplace = True)
test['TOP_PACK'].fillna(test['TOP_PACK'].mode()[0], inplace = True)
```

```python
encoding = train.groupby('TOP_PACK').size()
encoding = encoding/len(train)
train['TOP_PACK'] = train.TOP_PACK.map(encoding)

encoding = test.groupby('TOP_PACK').size()
encoding = encoding/len(test)
test['TOP_PACK'] = test.TOP_PACK.map(encoding)
```

```python
dummy1 = pd.get_dummies(train['REGION'], prefix_sep = '_')
dummy2 = pd.get_dummies(test['REGION'], prefix_sep = '_')

dummy3 = pd.get_dummies(train['TENURE'], prefix_sep = '_')
dummy4 = pd.get_dummies(test['TENURE'], prefix_sep = '_')

# dummy5 = pd.get_dummies(train['TOP_PACK'], prefix_sep = '_')
# dummy6 = pd.get_dummies(test['TOP_PACK'], prefix_sep = '_')
```

```python
train = pd.concat([train, dummy1, dummy3], axis = 1)
test = pd.concat([test, dummy2, dummy4], axis = 1)

train.drop(['REGION','TENURE'], axis = 1, inplace =True)
test.drop(['REGION','TENURE'], axis = 1, inplace =True)
```

```python
data = train.drop('CHURN', axis =1)
target = train['CHURN']
```
