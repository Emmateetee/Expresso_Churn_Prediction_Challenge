```python
dummy1 = pd.get_dummies(train['REGION'], prefix_sep = '_')
dummy2 = pd.get_dummies(test['REGION'], prefix_sep = '_')

dummy3 = pd.get_dummies(train['TENURE'], prefix_sep = '_')
dummy4 = pd.get_dummies(test['TENURE'], prefix_sep = '_')

dummy5 = pd.get_dummies(train['TOP_PACK'], prefix_sep = '_')
dummy6 = pd.get_dummies(test['TOP_PACK'], prefix_sep = '_')
```

```python
train = pd.concat([train, dummy1, dummy3, dummy5], axis = 1)
test = pd.concat([test, dummy2, dummy4, dummy6], axis = 1)

train.drop(['REGION','TENURE','TOP_PACK'], axis = 1, inplace =True)
test.drop(['REGION','TENURE', 'TOP_PACK'], axis = 1, inplace =True)
```

```python
data = train.drop('CHURN', axis =1)
target = train['CHURN']
```

```python
from sklearn.preprocessing import scale
```

```python
data_scaled = scale(data)
test_scaled = scale(test)
```

```python
data_ = pd.DataFrame(data_scaled, columns = data.columns)
test_ = pd.DataFrame(test_scaled, columns = test.columns)
```

```python
# IMPORT THE CATBOOST CLASSIFIER
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,log_loss, confusion_matrix
from sklearn.model_selection import StratifiedKFold
```

```python
kf = StratifiedKFold(n_splits=10, random_state= 42)
```

```python
# INITIALIZE THE MODEL
cat = CatBoostClassifier(learning_rate = 0.1, max_depth = 5, n_estimators = 5000, 
                         eval_metric = 'Accuracy',reg_lambda=50,random_state=101)
```

```python
scores = []
test_scores = np.zeros(len(test_))
for fold,(tr_in,te_in) in enumerate(kf.split(data_, target)):
    print(f"================================Fold{fold}====================================")
    X_train,X_test = data_.iloc[tr_in],data_.iloc[te_in]
    y_train,y_test = target.iloc[tr_in],target.iloc[te_in]
    cat.fit(X_train,y_train,eval_set=[(X_train,y_train),(X_test,y_test)],early_stopping_rounds=500,use_best_model=True)
    y_pred = cat.predict_proba(X_test)[:,1]
    scores.append(accuracy_score(y_test,cat.predict(X_test)))
    test_scores += cat.predict_proba(test_)[:,1]
```
