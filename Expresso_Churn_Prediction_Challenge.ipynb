```python
train.drop('user_id',axis = 1, inplace = True)
test.drop('user_id',axis = 1, inplace = True)
```

```python
train['REGION'].fillna(train['REGION'].mode()[0], inplace = True)
test['REGION'].fillna(test['REGION'].mode()[0], inplace = True)
```

```python
train['MONTANT'].fillna(0, inplace = True)
test['MONTANT'].fillna(0, inplace = True)

train['MONTANT'] = np.log1p(train['MONTANT'])
test['MONTANT'] = np.log1p(test['MONTANT'])
```

```python
train['FREQUENCE_RECH'].fillna(0, inplace = True)
test['FREQUENCE_RECH'].fillna(0, inplace = True)

train['FREQUENCE_RECH'] = np.log1p(train['FREQUENCE_RECH'])
test['FREQUENCE_RECH'] = np.log1p(test['FREQUENCE_RECH'])
```

```python
train['REVENUE'].fillna(0, inplace = True)
test['REVENUE'].fillna(0, inplace = True)

train['REVENUE'] = np.log1p(train['REVENUE'])
test['REVENUE'] = np.log1p(test['REVENUE'])
```

```python
train['ARPU_SEGMENT'].fillna(0, inplace = True)
test['ARPU_SEGMENT'].fillna(0, inplace = True)

train['ARPU_SEGMENT'] = np.log1p(train['ARPU_SEGMENT'])
test['ARPU_SEGMENT'] = np.log1p(test['ARPU_SEGMENT'])
```

```python
train['FREQUENCE'].fillna(0, inplace = True)
test['FREQUENCE'].fillna(0, inplace = True)

train['FREQUENCE'] = np.log1p(train['FREQUENCE'])
test['FREQUENCE'] = np.log1p(test['FREQUENCE'])
```

```python
train['DATA_VOLUME'].fillna(0, inplace = True)
test['DATA_VOLUME'].fillna(0, inplace = True)

train['DATA_VOLUME'] = np.log1p(train['DATA_VOLUME'])
test['DATA_VOLUME'] = np.log1p(test['DATA_VOLUME'])
```

```python
train['ON_NET'].fillna(0, inplace = True)
test['ON_NET'].fillna(0, inplace = True)

train['ON_NET'] = np.log1p(train['ON_NET'])
test['ON_NET'] = np.log1p(test['ON_NET'])
```

```python
train['ORANGE'].fillna(0, inplace = True)
test['ORANGE'].fillna(0, inplace = True)

train['ORANGE'] = np.log1p(train['ORANGE'])
test['ORANGE'] = np.log1p(test['ORANGE'])
```

```python
train['TIGO'].fillna(0, inplace = True)
test['TIGO'].fillna(0, inplace = True)

train['TIGO'] = np.log1p(train['TIGO'])
test['TIGO'] = np.log1p(test['TIGO'])
```

```python
train['ZONE1'].fillna(0, inplace = True)
test['ZONE1'].fillna(0, inplace = True)

train['ZONE1'] = np.log1p(train['ZONE1'])
test['ZONE1'] = np.log1p(test['ZONE1'])
```

```python
train['ZONE2'].fillna(0, inplace = True)
test['ZONE2'].fillna(0, inplace = True)

train['ZONE2'] = np.log1p(train['ZONE2'])
test['ZONE2'] = np.log1p(test['ZONE2'])
```

```python
train.drop('MRG',axis = 1, inplace = True)
test.drop('MRG',axis = 1, inplace = True)
```

```python
train['FREQ_TOP_PACK'].fillna(0, inplace = True)
test['FREQ_TOP_PACK'].fillna(0, inplace = True)

train['FREQ_TOP_PACK'] = np.log1p(train['FREQ_TOP_PACK'])
test['FREQ_TOP_PACK'] = np.log1p(test['FREQ_TOP_PACK'])
```

```python
train['TOP_PACK'].fillna(train['TOP_PACK'].mode()[0], inplace = True)
test['TOP_PACK'].fillna(test['TOP_PACK'].mode()[0], inplace = True)
```

```python
# from sklearn.preprocessing import LabelEncoder
# train['TOP_PACK'] = LabelEncoder().fit_transform(train['TOP_PACK'])
# test['TOP_PACK'] = LabelEncoder().fit_transform(test['TOP_PACK'])
```

```python
encoding = train.groupby('TOP_PACK').size()
encoding = encoding/len(train)
train['TOP_PACK'] = train.TOP_PACK.map(encoding)

encoding = test.groupby('TOP_PACK').size()
encoding = encoding/len(test)
test['TOP_PACK'] = test.TOP_PACK.map(encoding)
```

```python
data = train.drop('CHURN', axis =1)
target = train['CHURN']
```

```python
# IMPORT THE CATBOOST CLASSIFIER
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score,log_loss, confusion_matrix
from sklearn.model_selection import StratifiedKFold
```

```python
kf = StratifiedKFold(n_splits=10, random_state= 42)
```

```python
# INITIALIZE THE MODEL
cat = CatBoostClassifier(learning_rate = 0.1, max_depth = 5, n_estimators = 5000, 
                         eval_metric = 'Accuracy',reg_lambda=50,random_state=101)
```

```python
scores = []
test_scores = np.zeros(len(test))
for fold,(tr_in,te_in) in enumerate(kf.split(data, target)):
    print(f"================================Fold{fold}====================================")
    X_train,X_test = data.iloc[tr_in],data.iloc[te_in]
    y_train,y_test = target.iloc[tr_in],target.iloc[te_in]
    cat.fit(X_train,y_train,cat_features= [0,1],eval_set=[(X_train,y_train),(X_test,y_test)],early_stopping_rounds=500,use_best_model=True)
    y_pred = cat.predict_proba(X_test)[:,1]
    scores.append(accuracy_score(y_test,cat.predict(X_test)))
    test_scores += cat.predict_proba(test)[:,1]
```
